```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('save/CliffWalk.pkl', 'rb') as fp:
    unpickler = pickle.Unpickler(fp)
    CW_var = unpickler.load()
x=CW_var['E']
y=CW_var['t']
plt.figure(figsize=(6,5))
plt.plot(x,y)
plt.ylabel('Episode Length',fontsize=20)
plt.xlabel('Episode',fontsize=20)
plt.axhline(y=12, linewidth=4, color='r')
plt.tight_layout()
plt.savefig('Cliff_walk.png',dpi=100, background='Transparent')
plt.show()
```
